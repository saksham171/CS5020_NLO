```python
import numpy as np
import random
import matplotlib.pyplot as plt

A1= []
for i in range(0,10):
    n = random.randint(1,10)
    A1.append(n)

A1 = np.array(A1)

B1= []
for i in range(0,10):
    n = round(random.randint(1,10)*random.random(),2)
    B1.append(n)

B1 = np.array(B1)

A2= []
for i in range(0,10):
    n = random.randint(1,20)
    A2.append(n)

A2 = np.array(A2)

B2= []
for i in range(0,10):
    n = round(random.randint(1,20)*random.random(),2)
    B2.append(n)

B2 = np.array(B2)


A3= []
for i in range(0,10):
    n = random.randint(1,30)
    A3.append(n)

A3 = np.array(A3)

B3= []
for i in range(0,10):
    n = round(random.randint(1,30)*random.random(),2)
    B3.append(n)

B3 = np.array(B3)


A4= []
for i in range(0,10):
    n = random.randint(1,40)
    A4.append(n)

A4 = np.array(A4)

B4= []
for i in range(0,10):
    n = round(random.randint(1,40)*random.random(),2)
    B4.append(n)

B4 = np.array(B4)


A5= []
for i in range(0,10):
    n = random.randint(1,50)
    A5.append(n)

A5 = np.array(A5)

B5= []
for i in range(0,10):
    n = round(random.randint(1,50)*random.random(),2)
    B5.append(n)

B5 = np.array(B5)


def func_call_A(A,B):

 
    # Plot without constraints
    
    x = np.arange(0,10,1)
    y = np.arange(0,10,1)
    [X,Y] = np.meshgrid(x,y)
    E = ((X*A)+Y-B)**2
    
    fig = plt.figure(1)
    ax = plt.axes(projection='3d')
    surf = ax.plot_surface(X, Y, E, cmap = plt.cm.cividis)      # surface plot
    ax.set_title(' a) No constraints ')
    ax.set_xlabel("x(1)")
    ax.set_ylabel("x(2)")
    ax.set_zlabel("f(x)")

    plt.show()

    
def func_call_B(A,B):
    
    
    # Plot with Positive constraints

    x = np.arange(0,10,1)
    y = np.arange(0,10,1)
    [X,Y] = np.meshgrid(x,y)
    E = ((X*A)+Y-B)**2
    EXP =  (X*A)+Y-B
    EP = np.where(EXP>0,EXP,0)       #  Positive constraint
    
    fig = plt.figure(2)
    ax = plt.axes(projection='3d')
    surf = ax.plot_surface(X, Y, EP, cmap = plt.cm.cividis)      # surface plot
    ax.set_title(' b) Positive constraint ')
    ax.set_xlabel("x(1)")
    ax.set_ylabel("x(2)")
    ax.set_zlabel("f(x)")

    plt.show()
    
def func_call_C(A,B):

    # plot with Negative constraints
    
    x = np.arange(0,10,1)
    y = np.arange(0,10,1)
    [X,Y] = np.meshgrid(x,y)
    E = ((X*A)+Y-B)**2
    EXP =  (X*A)+Y-B
    EN = np.where(EXP<0,EXP,0)       #  Negative constraint

    fig = plt.figure(3)
    ax = plt.axes(projection='3d')
    surf = ax.plot_surface(X, Y, EN, cmap = plt.cm.cividis)     # surface plot
    ax.set_title(' c) Negative constraint')
    ax.set_xlabel("x(1)")
    ax.set_ylabel("x(2)")
    ax.set_zlabel("f(x)")

    plt.show()

func_call_A(A1,B1)
func_call_B(A1,B1)
func_call_C(A1,B1)

func_call_A(A2,B2)
func_call_B(A2,B2)
func_call_C(A2,B2)

func_call_A(A3,B3)
func_call_B(A3,B3)
func_call_C(A3,B3)

func_call_A(A4,B4)
func_call_B(A4,B4)
func_call_C(A4,B4)

func_call_A(A5,B5)
func_call_B(A5,B5)
func_call_C(A5,B5)
```
