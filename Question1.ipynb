```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-20, 20, 30)
y = np.linspace(-20, 20, 30)

def func(x,y) :
  z = 0.1*(x-1)*(x-1)*(x-1)*(x-1) + 2*x*y + 4*x*x -16*x -13*y  + 0.9*y*y*y*y + 4*y*y + 24
  return z

X, Y = np.meshgrid(x, y)
Z = func(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_title('Plot of the function f(x(1), x(2)) ')
ax.set_xlabel('x(1)')
ax.set_ylabel('x(2)')
ax.set_zlabel('f(x(1), x(2))')

plt.show()
```
