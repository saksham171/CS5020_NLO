```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

def f(x1, x2): return 100*(x2-x1**2)**2+(x1-1)**2

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_title('the Rosenbrock function')
ax.set_xlabel('x(1)')
ax.set_ylabel('x(2)')

plt.show()
```
